## 7.5 光滑样条 （smoothing Splines）

- 为什么我们需要g(x)光滑？极端情况｜减少方差
- 如何保证 g(x)光滑？💡LASSO

我们的任务是寻找到g能够满足 $RSS=\sum_{i=1}^{n}{(y_i-g(x_i))^2}$ 同样使得曲线也异常光滑
所以我们选择

###       目标函数：损失函数+惩罚项

$$\sum_{i=1}^{n}{(y_i-g(x_i))^2}+\lambda \int g''(t)^2 \,{\rm d}t\tag{7.11}$$
#### REAMRK：
1. 二阶导数的作用？
2. $\lambda=0 $ & $ \lambda=\infty$?
3. 最小化公式在极端值的性质？

### 光滑参数 $\lambda$
- $\lambda$ 参数控制光滑样条的的粗糙程度，也控制着有效自由度(effect of freedom）,$\lambda$ 允许从0 增加到 $+\infty$, 实际自由度（$df_\lambda$）从n 降到2
- 自由度和有效自由度：自由度指的是自由参数个数。有效自由度用来衡量光滑样条的光滑性的，其值越大说明样条更光滑（低偏差高方差）。
- 有效自由度的数学定义：
令  $\hat{g}_\lambda=S_\lambda  y$表示光滑样条拟合向量可以写成n*n 矩阵 $S_\lambda$ 乘响应向量$y$ , 有效自由度定义：
$${df}_\lambda=\sum^{n}_{i=1}\{S_\lambda\}_{ii}$$
有效自由度是矩阵$S_\lambda\ $ 对角元素之和。
- 如何选择 $\lambda$ ？交叉验证最小化LOOCV 留一交叉验证误差

$$RSS(\lambda)=\sum^{n}_{i=1}[\frac{y_i-\hat{g}_\lambda(x_i)}{1-\{S_\lambda\}_{ii}}]^2
$$
### 样条Spline、自由度 、
- 学习这些概念的意义？帮助理解模型；参数调整；
- 再次复习一下基本的概念
    - **样条spline**：具有m-1个连续导数的m 阶分段多项式
    - **结点knot**：基函数临界发生变化的结点
    - **自由度**： 三次样条指的是具有一组约束（连续性、一阶和二阶连续性）的分段多项式。通常情况下，具有K个节点的三次样条一般有(K+1)×4-K×3，也就是K+4个维度。当K=3时，维度为8，自由度是维度-1=7。一般情况下，我们只用三次样条。
    - **自然样条**

```python
age_grid = np.arange(df.age.min(), df.age.max()).reshape(-1,1)
transformed_x = dmatrix("bs(df.age, knots=(25,40,60), degree=3, include_intercept=False)",
                        {"df.age": df.age}, return_type='dataframe')
fit4 = sm.GLM(df.wage, transformed_x).fit()
pred4 = fit4.predict(dmatrix("bs(age_grid, knots=(25,40,60), degree=3, include_intercept=False)",
                             {"age_grid": age_grid}, return_type='dataframe'))
fit4.params
```

```python
# 6个自由度
transformed_x2 = dmatrix("bs(df.age, df=6, degree=3, include_intercept=False)",
                        {"df.age": df.age}, return_type='dataframe')
fit5 = sm.GLM(df.wage, transformed_x2).fit()
pred5 = fit5.predict(dmatrix("bs(age_grid, df=6, degree=3, include_intercept=False)",
                             {"age_grid": age_grid}, return_type='dataframe'))
fit5.params
```

```python
# 4个自由度
transformed_x3 = dmatrix("cr(df.age, df=4)", {"df.age": df.age}, return_type='dataframe')
fit6 = sm.GLM(df.wage, transformed_x3).fit()
pred6 = fit6.predict(dmatrix("cr(age_grid, df=4)", {"age_grid": age_grid}, return_type='dataframe'))
fit6.params
```

### 算法实现：
1. 选取占所有数据$s=k/n$比例最靠近$x_0$的数据$x_j$
2. 对选出点赋予权重$ K_{i0}=K(x_i,x_0)$进行计算
3. 最小化加权最小二乘回归
4. 根据$\hat{f}(x_0)=\hat{\beta_0}+\hat{\beta_1}x_0$得到$x_0$拟合结果



# 7.7 广义可加模型（generalized additive model GAM）

广义可加模型（generalized additive model GAM）提供了一种对标准线性进行推广的框架。这个框架中，每个变量用一个非线形函数替换，同时保持模型可加性。
## 回归问题
使用光滑的非线形函数$f_j(x_{ij})$代替$\beta_j x_{ij}$
即：
$$y=\beta_0+\beta_1f(X_1)+\beta_2f(X_2)+\dots+\beta_pf(X_p)+\varepsilon_i$$

### GAM 优点
1. 允许每个$x_j$ 拟合非线形 $f_j$,自动对标准的线形回归模型所忽略的非线形关系建模
2. 非线性模型将响应变量预测的更加精确
3. 可分析单个自变量对于响应变量的影响效果分析效果和自由度分析。

### GAM 缺点：
1.由于可加形式，会忽略有意义的交互项。但是可以增加$X_i * X_j$的交互项进行修正


想要摆脱上面的限定，可以采用随机森林和提升法之类更加光滑的方法。GAM 可以被看作是线形模型和完全非参数模型之间一类折中的方法


## 分类问题
GAM logistic 模型：

$$log (\frac{p(X)}{1-p(X)} )=\beta_0+f(X_1)+f(X_2)+\dots +f(X_p)$$

```python
# 局部回归 span = [0.1, 1, 10, 100]
# 刚开始没找到合适的函数 自己写了一个类
from sklearn import neighbors
import sklearn.linear_model as skl_lm
from sklearn.utils.validation import check_array
from sklearn.neighbors._base import _get_weights, _check_weights
import numpy as np


class LocalRegressor(neighbors.KNeighborsRegressor):
    
    def predict(self, X):
      
        X = check_array(X, accept_sparse='csr')

        # NN of X with respect to the train data (_fit_X)
        neigh_dist, neigh_ind = self.kneighbors(X)

        weights = _get_weights(neigh_dist, self.weights)
        
        _y = self._y
        _fit_X = self._fit_X
        if _y.ndim == 1:
            _y = _y.reshape((-1, 1))

        y_pred = np.empty((X.shape[0], 1), dtype=np.float64)

        # NN in the train data
        X_nn = np.squeeze(_fit_X[neigh_ind])
        y_nn = np.squeeze(_y[neigh_ind])
        
        # loop over the samples, not ideal from a speed point of view
        for i in range(X_nn.shape[0]):
            linear = skl_lm.LinearRegression()
            if weights is not None:
                linear.fit(X_nn[i].reshape(-1, 1), y_nn[i], sample_weight=weights[i])
            else:
                linear.fit(X_nn[i].reshape(-1, 1), y_nn[i])
            y_pred[i] = linear.predict(X[i].reshape(-1, 1))

        if self._y.ndim == 1:
            y_pred = y_pred.ravel()

        return y_pred

n_neighbors = len(X)  
sigmas = [0.1, 1, 10, 100]
y_preds = []
for sigma in sigmas:
    kernel = lambda dist: np.exp(-dist**2/(2*sigma**2))
    local_regression = LocalRegressor(n_neighbors, weights=kernel)
    local_regression.fit(X, y)
    y_preds.append(local_regression.predict(age_range))
fig, ax = plt.subplots(1, 1, figsize=(5,4))

ax.scatter(X, y, facecolors='None', edgecolors='grey', alpha=.5)
for num, sigma in enumerate(sigmas):
    ax.plot(age_range, y_preds[num], label=f'sigma: {sigma}')
ax.legend();
ax.set_xlabel('age')
ax.set_ylabel('wage');
```
