```python
left = pd.DataFrame({'key': ['foo', 'foo'], 'lval': [1, 2]})
right = pd.DataFrame({'key': ['foo', 'foo'], 'rval': [4, 5]})#设置两个dataframe
pd.merge(left, right, on='key')#依靠key列相同的指标进行枚举式的加列，其中left，right决定了哪些数据在前面
```

```python
left = pd.DataFrame({'key': ['foo', 'bar'], 'lval': [1, 2]})
right = pd.DataFrame({'key': ['foo', 'bar'], 'rval': [4, 5]})
pd.merge(left, right, on='key')#若有foo和bar 的区别，则foo和foo进行枚举，bar和bar进行枚举
```

```python
left = pd.DataFrame({'key': ['foo', 'bar'], 'lval': [1, 2]})
right = pd.DataFrame({'key': ['foo', 'nothing'], 'rval': [4, 5]})
pd.merge(left, right, on='key')#bar和nothing无法找到对应的匹配，则不会呗匹配出来
```

```python
rng = pd.date_range('1/1/2012', periods=120, freq='S')#S代表秒
ts = pd.Series(np.random.randint(0, 500, len(rng)), index=rng)#每秒生成一个随机数
ts.resample('1Min').sum()#resample表示重新采样，把每秒的数据转化成每分钟的，sum是求和，显然我们也可以用其他函数，比如mean
```
