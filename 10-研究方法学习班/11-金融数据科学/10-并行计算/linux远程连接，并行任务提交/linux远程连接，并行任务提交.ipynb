#### ls 命令，展示文件夹内内容

```python
-a ：全部的档案，连同隐藏档( 开头为 . 的档案) 一起列出来～ 
-A ：全部的档案，连同隐藏档，但不包括 . 与 .. 这两个目录，一起列出来～ 
-d ：仅列出目录本身，而不是列出目录内的档案数据 
-f ：直接列出结果，而不进行排序 (ls 预设会以档名排序！) 
-F ：根据档案、目录等信息，给予附加数据结构，例如： 
*：代表可执行档； /：代表目录； =：代表 socket 档案； |：代表 FIFO 档案； 
-h ：将档案容量以人类较易读的方式(例如 GB, KB 等等)列出来； 
-i ：列出 inode 位置，而非列出档案属性； 
-l ：长数据串行出，包含档案的属性等等数据； 
-n ：列出 UID 与 GID 而非使用者与群组的名称 (UID与GID会在账号管理提到！) 
-r ：将排序结果反向输出，例如：原本档名由小到大，反向则为由大到小； 
-R ：连同子目录内容一起列出来； 
-S ：以档案容量大小排序！ 
-t ：依时间排序 
```

## 4.并行计算任务提交

例子：用机器学习XGBoost算法，根据因子预测收益，首先需要对XGBoost算法进行超参数调整。由于因子和收益序列有时间先后，不能直接用GridSearchCV方法进行截面调参，本例子仅以一期参数调整为例，说明参数调整过程。用第一年前11个月的数据训练，第12个月的数据验证，获得用于以后月度预测的超参数。
- gridsingle.py文件 单进程一期调参

```python
from xgboost.sklearn import XGBRegressor
import datetime
import pandas as pd
import numpy as np
import time
from sklearn.metrics import mean_absolute_error
train=pd.read_csv('train.csv',index_col=0)
vali=pd.read_csv('test.csv',index_col=0)
x_train=np.array(train.iloc[:,:-1])
y_train=np.array(train.iloc[:,-1])
x_vali=np.array(vali.iloc[:,:-1])
y_vali=np.array(vali.iloc[:,-1])
start_t = datetime.datetime.now()
maemin = 100
for ne in [50,100,150,200,250]:
    print(ne)
    for lr in [0.05,0.1,0.15,0.2,0.25,0.3]:
        for md in [1,2,3,4,5,6,7]:
            # time.sleep(0.5)
            clf=XGBRegressor(n_estimators=ne, max_depth=md,
                         learning_rate=lr, n_jobs=4)
            clf.fit(x_train,y_train)
            PREDICTION = clf.predict(x_vali)
            mae = mean_absolute_error(PREDICTION, y_vali)
            if mae<maemin:
                maemin=mae
                lrb=lr
                mdb=md
                neb=ne
print([maemin,lrb,mdb,neb])
end_t = datetime.datetime.now()
elapsed_sec = (end_t - start_t).total_seconds()
print("单进程计算 共消耗: " + "{:.2f}".format(elapsed_sec) + " 秒")
```

- grid_multi.py文件 多进程一期调参

```python
from xgboost.sklearn import XGBRegressor
import datetime
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error
import multiprocessing as mp    ###多进程依赖的包


train=pd.read_csv('train.csv',index_col=0)
vali=pd.read_csv('test.csv',index_col=0)
x_train=np.array(train.iloc[:,:-1])
y_train=np.array(train.iloc[:,-1])
x_vali=np.array(vali.iloc[:,:-1])
y_vali=np.array(vali.iloc[:,-1])

##把要进行多线程的计算部分定义成函数形式
def f(v,name,ne,l):
    l.acquire()
    for lr in [0.05,0.1,0.15,0.2,0.25,0.3]:
        for md in [1,2,3,4,5,6,7]:
            clf=XGBRegressor(n_estimators=ne, max_depth=md,
                         learning_rate=lr, n_jobs=4)
            clf.fit(x_train,y_train)
            PREDICTION = clf.predict(x_vali)
            mae = mean_absolute_error(PREDICTION, y_vali)
            if mae<v[0]:
                v[0]=mae
                v[1]=lr
                v[2]=md
                v[3]=ne
                print(list(v))
#     l.acquire()                
    l.release()
    return {name: [v[1], v[2], v[3]]}


def main():
    start_t = datetime.datetime.now()
    l = mp.Lock()                    ##加锁可以保证多个进程修改同一块数据时，同一时间只能有一个任务可以进行修改，即串行的修改，
                                     # 速度是慢了，但牺牲了速度却保证了数据安全。
    v = mp.Array('d', [100,0,0,0])   ##设置共享值，用于获取最优超参数
    num_cores = int(mp.cpu_count())  ##计算机的核数
    print("计算机有: " + str(num_cores) + " 核心")
    pool = mp.Pool(num_cores-4)      ##指定多进程运行的核数
    pol = []
    neset = {'task1': 50,'task2': 100,'task3': 150, 'task4': 200, 'task5': 250}##
    for name, ne in neset.items():
        print(ne)
        p = mp.Process(target=f, args=(v,name,ne,l,))  ###只能一批一批地添加进程，同一批次内并行。参数后面一定要加逗号
        p.start()    #启动进程活动
        pol.append(p)
    for p in pol:
        p.join()   #使主调进程（包含XXX.join()语句的进程）阻塞，直至被调用进程XXX运行结束或超时（如指定timeout）
    print(list(v))
    end_t = datetime.datetime.now()
    elapsed_sec = (end_t - start_t).total_seconds()
    print("多进程计算 共消耗: " + "{:.2f}".format(elapsed_sec) + " 秒")

if __name__ == '__main__':
    main()
```

例子：用机器学习XGBoost算法，根据因子预测收益，首先需要对XGBoost算法进行超参数调整。由于因子和收益序列有时间先后，不能直接用GridSearchCV方法进行截面调参，本例子仅以滑动窗口每年参数调整为例，说明参数调整过程。每年用前11个月的数据训练，第12个月的数据验证，获得用于下一年月度预测的超参数。
- gridsingle2.py  单进程多期调参

```python
import glob, os
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error
from xgboost.sklearn import XGBRegressor
import datetime

# 读取因子数据和收益文件。因子默认使用因子库中的前10个。
# os.chdir(r'D:\financialengineer\parallel')
path = 'factor16'
file = glob.glob(os.path.join(path, "*.csv"))
k = []  ##存放因子+超额收益
factor = []  ##存放因子名称
for i in range(len(file)):
    k.append(pd.read_csv(file[i], header=0, index_col=0))
    factor.append(file[i][9:-4])
ret = pd.read_csv('exret30.csv', header=0, index_col=0)  # 股票月度收益
k.append(ret)
factor.append('exret')

# *****对原始数据进行预处理，每个截面一个Dataframe#
timeseries = []
for i in range(len(ret.index)):  # i=0#len(ret.index)
    kl = pd.concat([k[j].iloc[i, :] for j in range(len(factor))], axis=1, join="outer")
    kl.columns = factor
    kl = kl[kl['exret'].notna()]
    kl = kl.fillna(0)
    timeseries.append(kl)
# 训练集和验证集
# 每年更新一次超参数的选择

start_t = datetime.datetime.now()
months=[]
lrbest = []
mdbest = []
nebest = []
meaminset=[]
train_len = 11
for i in range(0, len(timeseries) - train_len, 12):  # i=264
    print(i)
    # 这里为了提高运算速度，将步长设置为12个月，也即每年调参一次。在每年年末获得新的最佳参数。用于下一年度的收益预测。
    # 因此，最早能够对收益做出预测的是1998年1月。投资组合也从这里开始。
    # 我们获得的最佳参数时间点分别为199712、199812、199912、···
    train = pd.concat(timeseries[i:train_len + i], axis=0)
    x_train = train.iloc[:, :len(factor) - 1]  # 因为一共选择了10个因子
    y_train = train.iloc[:, len(factor) - 1]  # 最后一列是收益

    validation = timeseries[train_len + i]
    x_vali = validation.iloc[:, :len(factor) - 1]
    y_vali = validation.iloc[:, len(factor) - 1]

    x_train = np.array(x_train)
    y_train = np.array(y_train)
    x_vali = np.array(x_vali)
    y_vali = np.array(y_vali)
    meamin = 100  ##平均绝对误差
    for lr in [0.05, 0.1, 0.15]:
        for md in [1, 2, 3]:
            for ne in [10, 50, 100]:
                clf = XGBRegressor(n_estimators=ne, max_depth=md,
                                   learning_rate=lr, n_jobs=4)
                clf.fit(x_train, y_train)
                PREDICTION = clf.predict(x_vali)
                mea = mean_absolute_error(PREDICTION, y_vali)
                if mea < meamin:
                    meamin = mea
                    lrb = lr
                    mdb = md
                    neb = ne
    months.append(i)
    lrbest.append(lrb)
    mdbest.append(mdb)
    nebest.append(neb)
    meaminset.append(meamin)
best_params = pd.DataFrame({'months':months,'learning_rate': lrbest, 'max_depth': mdbest, 'n_estimators':nebest,'mea':meaminset})
best_params.to_csv('best_params2.csv')
print(best_params)
end_t = datetime.datetime.now()
elapsed_sec = (end_t - start_t).total_seconds()
print("单进程计算 共消耗: " + "{:.2f}".format(elapsed_sec) + " 秒")
```

- grid_multi2.py文件 多进程多期调参

```python
import glob, os
from multiprocessing.pool import ApplyResult
from typing import List, Any, Union
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error
from xgboost.sklearn import XGBRegressor
import datetime
import multiprocessing as mp  

# 读取因子数据和收益文件。因子默认使用因子库中的前10个。
# os.chdir(r'D:\financialengineer\parallel')
path = 'factor16'
file = glob.glob(os.path.join(path, "*.csv"))
k = []  ##存放因子+超额收益
factor = []  ##存放因子名称
for i in range(len(file)):
    k.append(pd.read_csv(file[i], header=0, index_col=0))
    factor.append(file[i][9:-4])
ret = pd.read_csv('exret30.csv', header=0, index_col=0)  # 股票月度收益
k.append(ret)
factor.append('exret')
# *****对原始数据进行预处理，每个截面一个Dataframe#
timeseries = []
for i in range(len(ret.index)):  # i=0#len(ret.index)
    kl = pd.concat([k[j].iloc[i, :] for j in range(len(factor))], axis=1, join="outer")
    kl.columns = factor
    kl = kl[kl['exret'].notna()]
    kl = kl.fillna(0)
    timeseries.append(kl)
# 训练集和验证集
# 每年更新一次超参数的选择
train_len = 11
def get_params(i,timeseries):
    # 这里为了提高运算速度，将步长设置为12个月，也即每年调参一次。在每年年末获得新的最佳参数。用于下一年度的收益预测。
    # 因此，最早能够对收益做出预测的是1998年1月。投资组合也从这里开始。
    # 我们获得的最佳参数时间点分别为199712、199812、199912、···
    print(i)
    train = pd.concat(timeseries[i:train_len + i], axis=0)
    x_train = train.iloc[:, :len(factor) - 1]  # 因为一共选择了10个因子
    y_train = train.iloc[:, len(factor) - 1]  # 最后一列是收益

    validation = timeseries[train_len + i]
    x_vali = validation.iloc[:, :len(factor) - 1]
    y_vali = validation.iloc[:, len(factor) - 1]

    x_train = np.array(x_train)
    y_train = np.array(y_train)
    x_vali = np.array(x_vali)
    y_vali = np.array(y_vali)
    res=[]
    meamin = 100  ##平均绝对误差
    for lr in [0.05, 0.1, 0.15]:
        for md in [1, 2, 3]:
            for ne in [10, 50, 100]:
                clf = XGBRegressor(n_estimators=ne, max_depth=md,
                                   learning_rate=lr, n_jobs=4)
                clf.fit(x_train, y_train)
                PREDICTION = clf.predict(x_vali)
                mea = mean_absolute_error(PREDICTION, y_vali)
                if mea < meamin:
                    meamin = mea
                    lrb = lr
                    mdb = md
                    neb = ne
    res.append(i)
    res.append(lrb)
    res.append(mdb)
    res.append(neb)
    res.append(meamin)
    return res

def main():
    start_t = datetime.datetime.now()
    num_cores = int(mp.cpu_count())     ##计算机的核数
    print("计算机有: " + str(num_cores) + " 核心")
    pool = mp.Pool(num_cores - 4)       ##指定多进程运行的核数
    ##异步执行指的是一批子进程并行执行，且子进程完成一个，就新开始一个，而不必等待同一批其他进程完成。
    results=[pool.apply_async(get_params, args=(i,timeseries)) for i in range(0, len(timeseries) - train_len, 12)]
    results = [p.get() for p in results]   ##取出计算结果
    best_params = pd.DataFrame(results)
    best_params.columns=['months','learning_rate','max_depth','n_estimators','mea']
    best_params.to_csv('best_params.csv')
    print(best_params)
    end_t = datetime.datetime.now()
    elapsed_sec = (end_t - start_t).total_seconds()
    print("多进程计算 共消耗: " + "{:.2f}".format(elapsed_sec) + " 秒")

if __name__ == '__main__':
    main()
```
