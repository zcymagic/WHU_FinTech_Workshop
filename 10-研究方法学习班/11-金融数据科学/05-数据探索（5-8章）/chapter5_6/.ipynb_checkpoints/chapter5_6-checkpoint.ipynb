### 5.2.2 逻辑运算符
python的bool运算比较奇怪，因为其定义了最基础的或与非运算符号or，and，not，但是你也可以使用|(或), &(与), ~(非)等符号进行计算。

* 首先python中的表示“非”的取反符号不是使用传统的“！”
* 虽然单词和符号在一定程度上会产生相同的效果，但是二者机制不尽相同。前者非not属于逻辑取反、后者~属于对位取反。对位取反机制如下如：
![title](img/对位取反.png)
* 而从用法来说，or、and和not只能对单个bool进行运算，不能对类似pd.Series格式数据进行操作（可以采用遍历操作），而符号|、&和~可以对bool型Series进行操作。
* 另外采用对位取反符号对单个bool型变量操作要小心出错，出现缺失值等形况也要小心…可能会有以下情况出现：


```python
x=(flights['month']==1)
x
```

```python
~x[0]
```

```python
~True
```

```python
#找出11月或者12月的航班
flights[(flights['month']==11) | (flights['month']==12)]
```

```python
#运用 .isin()方法
nov_dec=flights[flights['month'].isin([11,12])]
nov_dec
```

```python
#找出延误时间（到达或者出发）不多于2小时的航班
#方法1
flights[~((flights['arr_delay']>120) | (flights['dep_delay']>120))]
```

```python
#方法2
flights[(flights['arr_delay']<=120) | (flights['dep_delay']<=120)]
```

#### PS:

我门发现以上两种方法最终得出的行数并不相同，这是因为之前所讲的python中把缺失值的比较认为是False。所以在方法一取反后含有缺失值的行对应的条件就变为True,而方法二中含有缺失值的行所对应的条件依旧为False，即被排除。

```python
#对应dense_rank()
#dense_rank()比较麻烦因为python中没有对应的函数，而需要把缺失值单独提取出来，排序后再将缺失值插入原位
#如果有现成的函数，或者是更好的算法，可以帮忙修改一下……
def dense_rank(x,ascending=True):#x为一个Series
    loc=x[x.isna()].index.values
    x.dropna().sort_values(ascending=ascending)
    for i in loc:
        x[i]=np.nan
    return x
dense_rank(y)
'''
by Chenxi Zhang, 20201117:
1.rank()中有dense方法
y.rank(ascending=1,method='dense')
'''
```

```python
#对应cume_dist()
y.rank(method='max')/y.rank(method='max').max() 
```

## 5.6分组摘要

```python
#计算某一列的均值，若直接df.mean()则返回各个列的均值
#python中的df.mean()函数自动过滤了缺失值
flights[['dep_delay']].mean()
```

```python
#对应列的聚合可以运用命名聚合函数pd.NameAgg(column,aggfunc)，可参考第一次学习班代码。
#但是为了让代码看的不这么冗长，运用字典和列表以及元组也可以达到相同的效果。
by_day=flights.groupby(['year','month','day']).agg({'dep_delay':[('delay','mean')]})
by_day.head()
#by_day返回值是一个dataframe
```

```python
#通过分组和聚合函数最后行程的表是比较不规整的，其中表格的index是由分组确定的，实际上是一个包含三个元素的元组。
#（PS：dataframe的索引必须为不能变更（比较不专业）的数据结构，不能是list、array、series格式）
#在看列的标签，是一个两个元素的元组，我们只取最高层级的标签，通过droplevel()函数可以实现
#reset_index()可以把index转为列
by_day.columns=by_day.columns.droplevel(0)
by_day.reset_index()
```

### 5.6.3 计数

```python
#查看一下具有最长平均延误时间的飞机（通过机尾编号进行识别）
#很遗憾的是matplotlib和seaborn不能直接绘制geom_freqploy()图，若想绘制一摸一样的图可以调用plotnine
delays=not_cancelled.groupby('tailnum')['arr_delay'].agg([('delay','mean')]).reset_index()
fig,ax=plt.subplots(1,1)
sns.distplot(delays['delay'],bins=int(len(delays['delay'])/10),
             norm_hist=False)
plt.show()
```

接下来我们考虑这种数据的另一类变体，我们运用Lahman包中的数据，探究棒球击球手平均表现与击球次数之间的关系。
我们需要利用Lahman包中的Batting数据集来计算大联盟的每个棒球队员的打击率（安打数/打数）

```python
#首先我们要导入数据
f=open('Lahman/Batting.csv')
batting=pd.read_csv(f)
```

```python
#由于将不同列聚合到单一变量中比较麻烦（如果有直接的函数，建议分享一下），还是采用对单列聚合后再计算的方式。
batters=batting.groupby('playerID').agg({'H':[('ba_dividend','sum')],'AB':[('ab','sum')]})
batters.columns=batters.columns.droplevel(0)
batters['ba']=batters['ba_dividend']/batters['ab']
batters=batters.iloc[:,1:].reset_index()
batters.head()
```

```python
#筛选出ab>100的值
sns.lmplot('ab','ba',data=batters[batters['ab']>100],lowess=True,aspect=1.5,
           scatter_kws={'color':'k'})
plt.show()
```

当我们绘制出击球手能力（用击打率ba衡量）与击球机会数量（用打数ab衡量）之间的关系时可以发现：

* 数据点越多，聚合值的变动就越小
* 能力（ba）和击球机会数量（ab）之间存在正相关，这是因为球队会控制击球手的出场，而球队会优先选择最好的队员

### 5.6.4 常用的摘要函数

```python
#单纯的均值计数是远不够的，我们一般需要将聚合函数和逻辑筛选组合起来使用
#计算平均延误时间和平均正延误时间
def mean2(arr):
    return arr[arr>0].mean()
not_cancelled.groupby(['year','month','day'])['arr_delay'].agg([('avg_delay1','mean'),('avg_delay2',mean2)]).reset_index()
```

```python
#为什么到某些目的地的距离比到其他目的地更多变？
not_cancelled.groupby('dest')['distance'].agg([('distance_sd','std')]).reset_index().sort_values(by='distance_sd',ascending=False)
```

```python
#每天最早和最晚的航班何时出发？
not_cancelled.groupby(['year','month','day'])['dep_time'].agg([('first','min'),('last','max')]).reset_index()
```

```python
#找出每天最早出发和最晚出发的航班
#first和last返回结果与上表一致，大家可以自行查查相应的参数设置
not_cancelled.groupby(['year','month','day'])['dep_time'].agg([('first_dep','first'),('last_dep','last')]).reset_index()
```

```python
#我们需要找出每一天最早出发和最迟出发的两个班次飞机所有信息
df=not_cancelled.sort_values(['year','month','day','dep_time'])
df['r']=df.groupby(['year','month','day'])['dep_time'].rank(method='min',ascending=False)

#自定义函数，筛选dataframe某一列排序最大和最小的两行
def df_range(df,column='r'):
    range1=list([df[column].min(),df[column].max()])
    return df[df[column].isin(range1)]

#运用apply方法，其主要将整个dataframe分割（split）成几个子集，而后对每个子集运用目标函数，最后再聚合（combine）
res=df.groupby(['year','month','day']).apply(df_range)
res.index=res.index.droplevel([0,1,2])
res
'''
by Chenxi Zhang, 20201117:
1.筛选每组中排名最高最低行可简化
df = df.groupby(['year','month','day']).apply(lambda x:x[(x.r == x.r.min())|(x.r == x.r.max())])
'''
```

```python
#那个目的地有最多航空公司
not_cancelled.groupby('dest')['carrier'].agg([('carriers',lambda x :len(x.unique()))]).sort_values(by='carriers',ascending=False)
```

## 5.7分组新变量（和筛选器）

```python
#找出每个分组中最差的成员
def filter_sml(df,column='arr_delay'):
    #注意R中的rank函数默认相当于python中method='average'，若参数不对则返回表格不尽相同。
    return df[df[column].rank(method='average',ascending=False)<10]

flights_sml=flights.groupby(['year','month','day']).apply(filter_sml)
flights_sml.index=flights_sml.index.droplevel([0,1,2])
flights_sml
```

```python
#找出大于某个阈值的所有分组。
def n_365(df,column='dest'):
    if df[column].count()>365:
        return df
    else:
        #若不加这一行则返回同样大小子集的nan值。
        return pd.DataFrame()

popular_dests=flights.groupby('dest').apply(n_365)
popular_dests.index=popular_dests.index.droplevel(0)
#为了使结果跟表中的一致，对popular_dest进行排序，但是按理使应该对列进行排序的，为了简便只对前四列排序。
popular_dests.sort_values(['year','month','day','dep_time'])
'''
by Chenxi Zhang, 20201117:
1.筛选大于阈值分组可用filter()
popular_dests=flights.groupby('dest').filter(lambda x:len(x)>365)
'''
```

```python
#对数据进行标准化计算分组指标，计算到达每个目的地的延迟比例，首先我们要求出每个目的地的延迟总时间
pop_filter=popular_dests[popular_dests['arr_delay']>0].copy()
#构造映射
map_sum=pop_filter.groupby('dest')['arr_delay'].sum()
map_sum
```

```python
#运用map函数和Series映射创建新变量
pop_filter['sum']=pop_filter['dest'].map(map_sum)
pop_filter[['dest','sum']]
```

```python
#数据整理
pop_filter['prop_delay']=pop_filter['arr_delay']/pop_filter['sum']
pop_select=pop_filter[['year','month','day']+['dest','arr_delay','prop_delay']]
pop_select.sort_values(['year','month','day'])
```
