```python
#按照课本叙述完成winsorize操作
def winsorize(series,degree):
    down = series.quantile(degree)
    up = series.quantile(1-degree)
    seriesname = series.name
    x = pd.Series(np.where(series<down, down, series),name = seriesname)
    x = pd.Series(np.where(x>up, up, x),name = seriesname)
    return x

def winsorize_wholedata(data,degree=0.005):
    temp = data.apply(lambda x:winsorize(x,degree))
    return temp
```

```python
#定义一下person协方差和spearmann协方差的计算方法
def personcorr_calculator(data,dataname1,dataname2,degree=0.005):
    X = []
    for y in range(1988,2013):
        x = data[data['year'] == y]
        x = x[[dataname1,dataname2]]
        x = x.dropna(axis=0,how='any')
        temp = winsorize_wholedata(x,degree)
        person_corr = temp.corr().iloc[0,1]
        X.append(person_corr)
    X = pd.Series(X,index = range(1988,2013),name = 'pt('+dataname1+','+dataname2+')')
    return X

```

To calculate the Spearman rank correlation, one must first calculate the ranking
for each entity i on each of X and Y. We let xi,t be the rank of Xi,t calculated over
all entities that have valid values of both X and Y during period t. Thus, if entity i
has the lowest value of X, xi,t is 1. If entity i has the highest value of X, then xi,t is
nt. If there are multiple entities for which the value of X is the same, then each of
these entities is assigned a ranking equal to the average position of these entities in
the ordered list of the entities when sorted on the variable X. The rankings for Y are
calculated analogously and are denoted yi,t. It should be noted that when calculating
the Spearman rank correlation, the data should not be winsorized. For each entity i,
the difference between the entity’s ranking on X and it’s ranking on Y is defined as
di,t = xi,t − yi,t. Finally, the Spearman rank correlation between X and Y for period t
is calculated as

```python
def spearmancorr_calculator(data,dataname1,dataname2):
    X = []
    for y in range(1988,2013):
        x = data[data['year'] == y]
        x = x[[dataname1,dataname2]]
        x = x.dropna(axis=0,how='any')
        spearman_corr = x.corr(method = 'spearman').iloc[0,1]
        X.append(spearman_corr)
    X = pd.Series(X,index = range(1988,2013),name = 'ps('+dataname1+','+dataname2+')')
    return X
```

```python
#生成表格
def corr_df(data):
    x1_p = personcorr_calculator(data,'beta','Size',degree=0.005)
    x1_s = spearmancorr_calculator(data,'beta','Size')
    x2_p = personcorr_calculator(data,'beta','BM',degree=0.005)
    x2_s = spearmancorr_calculator(data,'beta','BM')
    x3_p = personcorr_calculator(data,'beta','rt_1',degree=0.005)
    x3_s = spearmancorr_calculator(data,'beta','rt_1')
    x4_p = personcorr_calculator(data,'Size','BM',degree=0.005)
    x4_s = spearmancorr_calculator(data,'Size','BM')
    x5_p = personcorr_calculator(data,'Size','rt_1',degree=0.005)
    x5_s = spearmancorr_calculator(data,'Size','rt_1')
    x6_p = personcorr_calculator(data,'BM','rt_1',degree=0.005)
    x6_s = spearmancorr_calculator(data,'BM','rt_1')
    x = pd.concat([x1_p,x1_s,x2_p,x2_s,x3_p,x3_s,x4_p,x4_s,x5_p,x5_s,x6_p,x6_s],axis=1)
    return x

dataframe_corr = corr_df(alldata)
data_corr = dataframe_corr.apply(lambda x:round(x, 2))#设为两位小数
data_corr
```

```python
corr_mean = dataframe_corr.mean().apply(lambda x:round(x, 2))
corr_mean=corr_mean.to_frame()
corr_mean
```
